## 1. My lane (or freestyle) and why

I chose **Lane 2: Refresh / Content Opportunity Scoring** because it focuses on a practical problem that content teams face. Websites usually contain hundreds or even thousands of pages, and reviewing each one manually takes a lot of time.

After exploring the starter dataset, I noticed that some pages appear to be losing performance while others haven't been updated for a long time. Instead of checking every page individually, I think it would be useful to build a system that ranks pages based on how urgently they need attention.

This project can help content teams decide which pages should be refreshed first, making the review process more organized and efficient.

## 2. The question: decision, action, cost of a wrong call

**Research question**

Which pages should be reviewed first because they are most likely to benefit from a content refresh?

**Decision**

The goal is to help a content team decide which pages deserve attention before spending time reviewing every page manually.

**Unit of analysis**

Each row in the dataset represents one web page.

**Output**

The model should produce a ranked list of pages along with a score showing how likely each page is to require a content refresh.

**Action**

A content writer or SEO specialist can start reviewing the pages with the highest scores first.

**Cost of a wrong recommendation**

If a healthy page is incorrectly flagged, some time may be wasted reviewing it.

If a page that actually needs attention is missed, it may continue losing visibility and traffic before anyone notices. Because of this, missing important pages is the more costly mistake.
